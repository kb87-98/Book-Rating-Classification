```python
df = pd.read_csv('/kaggle/input/testprepdata/TestPreprossedData.csv')
df.shape
```

```python
df.drop(df.iloc[:,0:2], inplace=True, axis=1)
df.drop(df.iloc[:,4:10],inplace=True,axis=1)
df
```

```python
x = df['review_text']
x
```

```python
x.isna().sum()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle

# Assuming 'df' is your DataFrame with the 'review_text' column already lemmatized

# Convert the lemmatized text back to strings
x.fillna('', inplace=True)

# Initialize the TF-IDF vectorizer
tfidf_vectorizer = pickle.load(open('/kaggle/input/modelandvec3/vec3.pickle', 'rb'))
xtfidf = tfidf_vectorizer.transform(x);
```

```python
pickled_model = pickle.load(open('/kaggle/input/modelandvec3/logistic_model_3.pkl', 'rb'))
ytest = pickled_model.predict(xtfidf)
```

```python
data = pd.DataFrame()
data['review_id'] = df['review_id']
data['rating'] = ytest
```
