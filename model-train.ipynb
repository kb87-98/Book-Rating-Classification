```python
df = pd.read_csv('/kaggle/input/ratemeter/train.csv')
df.shape
```

```python
df2 = pd.read_csv('/kaggle/input/ppdata/preprocessedAll.csv')
df2.shape
```

```python
df2.drop(df2.iloc[:,0:2], inplace=True, axis=1)
df2.drop(df2.iloc[:,4:10],inplace=True,axis=1)
```

```python
df2
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Assuming 'df' is your DataFrame with the 'review_text' column already lemmatized

# Convert the lemmatized text back to strings
df2['review_text'].fillna('', inplace=True)

# Initialize the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=20000,stop_words='english', ngram_range=(1,2))
```

```python
x=df['review_text']
```

```python
y=df['rating']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
#Now finally split the dataset
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
tfidf_vectorizer.fit(X_train)

# transform
tfidf_X_train = tfidf_vectorizer.fit_transform(X_train)
tfidf_X_test = tfidf_vectorizer.transform(X_test)
```

```python
import pickle

pickle.dump(tfidf_vectorizer, open("vec3.pickle", "wb"))
```

```python
from sklearn.linear_model import LogisticRegression
# from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
```

```python
# Hyperparameters
# C = [0.001, 0.01, 0.1, 1, 10]

# for c in C: 
# Define model
log_model = LogisticRegression(C=0.7, max_iter=1000)


_ = log_model.fit(tfidf_X_train, y_train)

# predictions
y_preds_train = log_model.predict(tfidf_X_train)
y_preds_test = log_model.predict(tfidf_X_test)

# evaluation
print()
print(log_model)
print(f"Train accuracy score : {accuracy_score(y_train, y_preds_train)}")
print(f"Test accuracy score : {accuracy_score(y_test, y_preds_test)}")

print('\n',40*'-')
```

```python
print(f"Test f1 score : {f1_score(y_test, y_preds_test, average='micro')}")
```
